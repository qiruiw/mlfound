```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_train = pd.read_csv("hw1_18_train.dat", header=None, delimiter=r"\s+", names=['x1','x2','x3','x4','y'])
data_train.head()
```

```python
data_test = pd.read_csv("hw1_18_test.dat", header=None, delimiter=r"\s+", names=['x1','x2','x3','x4','y'])
data_test.head()
```

```python
def get_x(data):
    data.insert(0, 'x0', 1)
    return np.array(data.iloc[:, :-1])

def get_y(data):
    return np.array(data.iloc[:, -1])
```

```python
X_train = get_x(data_train)
y_train = get_y(data_train)
print('shapes:\nX: {}\ny: {}'.format(X_train.shape, y_train.shape))
```

```python
w = np.zeros(X_train.shape[1])
```

```python
def signal(a):
    return 1 if a>0 else -1

signalize = np.vectorize(signal)
```

```python
def find_mistake(w, X, y, i, cycle):
    conflict = True
    i = (i+1) % y.shape[0]
    iterations = 0
    while conflict:
        indx = cycle[i]
        if signal(X[indx, :] @ w) != y[indx]:
            iterations = 0
            return i
        else:
            iterations += 1
        i = (i+1) % y.shape[0]
        if iterations == y.shape[0]:
            conflict = False
    return -1
    
def error(w, X, y):
    predict = signalize(X @ w)
    return (predict != y).astype(int).sum()/y.shape[0]

def pocket(w, X, y, it, eta=1, cycle=np.arange(400)):
    i = -1
    #elist=[];indlist=[]
    w_pocket = w
    for it in np.arange(it):
        i = find_mistake(w, X, y, i, cycle)
        if i == -1: # find no mistakes
            break
        indx = cycle[i]
        #indlist.append(indx)
        #print(w)
        w = w + eta*y[indx]*X[indx, :]
        
        curr_error = error(w, X, y)
        pocket_error = error(w_pocket, X, y)
        #elist.append(pocket_error)
        #print(curr_error, pocket_error)
        if curr_error < pocket_error:
            w_pocket = w
    
    #plt.figure()
    #plt.plot(elist)
    #plt.show()
    #print(indlist)   
    return w_pocket

def train_2000(w, X_train, y_train, X_test, y_test, it):
    total_error = 0
    for i in np.arange(2000):
        np.random.seed(np.random.randint(10000))
        cycle = np.random.permutation(y_train.shape[0])
        w_final = pocket(w, X_train, y_train, it, cycle=cycle)
        total_error += error(w_final, X_test, y_test)
        #total_update_counter += pla(w, X, y, 0.5, cycle)
    return total_error/2000
```

```python
pocket(w, X_train, y_train, 10)
```

```python
X_test = get_x(data_test)
y_test = get_y(data_test)
```

```python
train_2000(w, X_train, y_train, X_test, y_test, 50)
```

```python
# return w_50 instead of w_pocket
def pocket_modified(w, X, y, it, eta=1, cycle=np.arange(400)):
    i = -1
    for it in np.arange(it):
        i = find_mistake(w, X, y, i, cycle)
        if i == -1: # find no mistakes
            break
        indx = cycle[i]
        w = w + eta*y[indx]*X[indx, :]

    return w

def train_2000_modified(w, X_train, y_train, X_test, y_test, it):
    total_error = 0
    for i in np.arange(2000):
        np.random.seed(np.random.randint(10000))
        cycle = np.random.permutation(y_train.shape[0])
        w_final = pocket_modified(w, X_train, y_train, it, cycle=cycle)
        total_error += error(w_final, X_test, y_test)
        #total_update_counter += pla(w, X, y, 0.5, cycle)
    return total_error/2000
```

```python
train_2000_modified(w, X_train, y_train, X_test, y_test, 50)
```

```python
train_2000(w, X_train, y_train, X_test, y_test, 100)
```
